### Transform date into a date format. Create columns for Year, month and day

```python
df['Date']=pd.to_datetime(df['Date'])
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
df.head()
```

#### Replace values
State Bottle Cost and State Bottle Retail have a '$' sign. We need to remove that character and turn it into a numeric type (float)

```python
df['State Bottle Cost']=df['State Bottle Cost'].str.replace('$', '').astype(float)
df['State Bottle Retail']=df['State Bottle Retail'].str.replace('$', '').astype(float)
df.head()
```

#### Change data types

```python
df.info()
```

Store number, Category, Vendor number, Item number, Pack, Year, Month and Day will be turned into strings (we don't do any calculations with those features - other than counting)

```python
df = df.astype({"Store Number": str, 
                "Category": str,
                "Vendor Number": str,
                "Item Number": str,
                "Pack": str,
               "Year": str,
               "Month": str,
               "Day": str})
df.info()
```

# Merge tables
Zip codes in IOWA https://worldpopulationreview.com/zips/iowa

It includes the zip code, city, county and population. We could do some additional analysis based on it

```python
zipIowa= pd.read_csv("ZipIowa.csv")
zipIowa.head()
```

```python
zipIowa.info()
```

Change the zip code into a string

```python
zipIowa = zipIowa.astype({"zip": str})
```

```python
df1=pd.merge(df, zipIowa, how="left", left_on='Zip Code', right_on='zip')
df1.head()
```

#### What are the top cities based on total sales?

```python
df1[['city','Total Sales']].groupby(by="city").sum().sort_values('Total Sales', ascending=False).head(10).plot(kind='bar')
```
